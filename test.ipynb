```python
import pandas as pd 
from mlxtend.frequent_patterns import apriori, association_rules

pd.low_memory = False
file_path = 'F:/sem5/DWDM/lab/healthcare_noshows.csv' 
df = pd.read_csv(file_path)

df.head()

df_apriori = df[['Scholarship', 'Hipertension', 'Diabetes', 'Alcoholism', 'Handcap', 'SMS_received', 'Showed_up']]

df_apriori = df_apriori.replace({'True': 1, 'False': 0, True: 1, False: 0, 'Yes': 1, 'No': 0})

df_apriori = df_apriori.fillna(0)

df_apriori = df_apriori.astype(int)

frequent_itemsets = apriori(df_apriori, min_support=0.10, use_colnames=True)

print(frequent_itemsets)

rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.10)

rules
```
